```python
from os.path import join

import cv2
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['figure.figsize'] = 20, 20
```

```python
root = '../data/images/'
```

# Visualize validation dataset

```python
val_df = pd.read_csv('../data/val.csv', index_col=0)
val_df.shape
```

```python
for i in range(1, 41):
    plt.subplot(8, 10, 2 * i - 1)
    plt.xticks([]); plt.yticks([])
    img = cv2.imread(join(root, val_df.iloc[i]['cleared']))
    plt.imshow(img)
    plt.subplot(8, 10, 2 * i)
    plt.xticks([]); plt.yticks([])
    img = cv2.imread(join(root, val_df.iloc[i]['ribs']))
    plt.imshow(img)
```
